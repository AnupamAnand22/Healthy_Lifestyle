```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
df["Food preference"]=imputer.fit_transform(df["Food preference"].values.reshape(-1,1))[:,0]
```

```python
df.isnull().sum()
```

```python
df.drop(['ID1','ID2'], axis = 1, inplace = True)
```

```python
def split(string):
    words = string.split(' ')
    return words
words = df['Food preference'].apply(split)
df['words'] = words
df['DX1'] = 0
df['DX2'] = 0
df['DX3'] = 0
df['DX4'] = 0
df['DX5'] = 0
df['DX6'] = 0
```

```python
for i in range(len(df)):
    if('DX1' in df.words[i]):
        df.DX1[i] = 1
    if('DX2' in df.words[i]):
        df.DX2[i] = 1
    if('DX3' in df.words[i]):
        df.DX3[i] = 1
    if('DX4' in df.words[i]):
        df.DX4[i] = 1
    if('DX5' in df.words[i]):
        df.DX5[i] = 1
    if('DX6' in df.words[i]):
        df.DX6[i] = 1
```

```python
df.head()
```

```python
df.drop(['words', 'Food preference'], axis = 1, inplace = True)
```

```python
df.drop(['index'], axis = 1, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
objList = df.select_dtypes(include = "object").columns
print(objList)
```

```python
for i in objList:
    df[i] = le.fit_transform(df[i].astype(str))

df.head(10)
```

```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from imblearn.combine import SMOTEENN
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
x=df.drop('Healthy',axis=1)
```

```python
y=df['Healthy']
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_m,y,test_size=0.2,stratify=df.Healthy)
```

```python
model1=LogisticRegression()
```

```python
model1.fit(x_train,y_train)
```

```python
pred1=model1.predict(x_test)
```

```python
model1.score(x_test,y_test)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, pred1).ravel()
print('True negatives: ', tn, '\nFalse positives: ', fp, '\nFalse negatives: ', fn, '\nTrue Positives: ', tp)
```

```python
print(classification_report(y_test, pred1, labels=[0,1]))
```
